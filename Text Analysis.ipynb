```python
f = open(r"D:\PythonCode\Natural Language Processing(NLP)\microsoft news.txt", "rt")
news = f.read()
```

```python
news
```

```python
import re
news = re.split(r'\W+', news)
print(news)
```

```python
#To convert the words in lower case
news = [words.lower() for words in news]
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
news_stop = [word for word in news if not word in stopwords ]
```

# Stemming

```python
from nltk.stem.porter import *

porter = PorterStemmer()
stemmed = [porter.stem(word) for word in news_stop]
print(stemmed)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import WordPunctTokenizer

lemmatizer = WordNetLemmatizer()
lemma = [lemmatizer.lemmatize(word) for word in stemmed]
```

```python
print(lemma)
```

# Text Analysis

```python
#Word Cloud Generation
!pip install wordcloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
microsoft_news = ''
microsoft_news += ' '.join(news_stop) + ' ' 
```

```python
microsoft_news
```

```python
from nltk.tag import pos_tag

def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent

preprocess(microsoft_news)
```

```python
def wordcloud(data, title):
    cloud = WordCloud(width=400,
                      height=330,
                      max_words=150,
                      colormap='tab20c',
                      collocations=True).generate(data)
    plt.figure(figsize=(10,8))
    plt.imshow(cloud)
    plt.axis('off')
    plt.title(title, fontsize=13)
    plt.show()
    
wordcloud(microsoft_news,'Microsoft News Analysis')
```

```python
import numpy as np
!pip install Pillow
import PIL
from PIL import Image
from numpy import asarray
image = Image.open(r'D:\PythonCode\Natural Language Processing(NLP)\Microsoft.png')
microsoft_image = asarray(image)
```

```python
def news_wordcloud(data, title, mask=None):
    cloud = WordCloud(scale=3,
                      colormap='RdYlGn',
                      mask=mask,
                      background_color='white',
                      collocations=True,
                      contour_color='black',
                      contour_width=1).generate(data)
    plt.figure(figsize=(10,8))
    plt.imshow(cloud)
    plt.axis('off')
    plt.title(title)
    plt.show()
    
news_wordcloud(microsoft_news, "Microsoft News Analysis", microsoft_image)
```

# Frequency of Words used in Microsoft News Analysis

```python
nlp_words=nltk.FreqDist(news_stop)
top10 ={k: nlp_words[k] for k in list(nlp_words)[:10]}
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca()
plt.bar(range(len(top10)), list(top10.values()), align='center')
plt.xticks(range(len(top10)), list(top10.keys()))
plt.show()
```
